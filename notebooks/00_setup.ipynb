```python
# Cell 5 — BM25 smoke test
import bm25s
corpus = ["a cat sat on the mat", "dogs chase cats", "the mat is red"]
tokens = bm25s.tokenize(corpus)
idx = bm25s.BM25()
idx.index(tokens)
res, scores = idx.retrieve(bm25s.tokenize(["cat"]), k=2)
print('results:', res)
print('scores:', scores)
```
